```python
import pandas as pd
import numpy as np

# Define the columns for the dataframe
columns = ['student_id', 'CGPA', 'IELTS', 'LOR', 'research_experience', 'work_experience', 'course', 'chances', 'university']

# Create an empty dataframe with the specified columns
df = pd.DataFrame(columns=columns)

# Generate 100 rows of mock data
for i in range(100):
    student_id = i
    CGPA = np.random.uniform(low=0.0, high=10.0)
    IELTS = np.random.randint(low=0, high=9)
#     GRE = np.random.randint(low=200, high=990)
    research_experience = np.random.choice([1, 0])
    work_experience = np.random.choice([1, 0])
    course = np.random.choice([1])
    LOR = np.random.randint(low=0, high=5)
    university = np.random.choice(['QMU'])
    chances = np.random.uniform(low=0.0, high=1.0,)
    
    # Add the generated values to the dataframe
    df = df.append({'student_id': student_id, 'CGPA': CGPA, 'IELTS': IELTS, 'research_experience': research_experience, 'work_experience': work_experience, 'course': course, 'LOR': LOR,'university': university, 'chances': chances}, ignore_index=True)

# Export the dataframe to a CSV file
df.to_csv('QMU.csv', index=False)

```
